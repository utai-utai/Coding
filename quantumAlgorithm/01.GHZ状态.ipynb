1. е®ҡд№үпјҡGHZ жҖҒпјҲGreenberger-Horne-Zeilinger stateпјүжҳҜдёҖз§ҚеӨҡдҪ“жңҖеӨ§зә зј жҖҒгҖӮеҜ№дәҺ $n$ дёӘйҮҸеӯҗдҪҚпјҢе…¶ж ҮеҮҶеҪўејҸдёәпјҡ$$|\text{GHZ}\rangle = \frac{|0\rangle^{\otimes n} + |1\rangle^{\otimes n}}{\sqrt{2}}$$еҜ№дәҺ 3 дёӘйҮҸеӯҗдҪҚпјҢе®ғжҳҜпјҡ$|\text{GHZ}\rangle = \frac{1}{\sqrt{2}}(|000\rangle + |111\rangle)$гҖӮ
2. ж ёеҝғзү№жҖ§пјҡжңҖеӨ§зә зј пјҡ зі»з»ҹдёӯд»»ж„ҸдёҖдёӘйҮҸеӯҗдҪҚзҡ„жөӢйҮҸз»“жһңйғҪдјҡзһ¬й—ҙеҶіе®ҡе…¶д»–жүҖжңүйҮҸеӯҗдҪҚзҡ„зҠ¶жҖҒгҖӮйқһеұҖеҹҹжҖ§йӘҢиҜҒпјҡ е®ғеёёиў«з”ЁжқҘиҜҒжҳҺйҮҸеӯҗеҠӣеӯҰдёҺиҙқе°”йқһеұҖеҹҹжҖ§пјҲBell's non-localityпјүпјҢжҜ”дј з»ҹзҡ„иҙқе°”жҖҒпјҲBell statesпјүеңЁжөӢиҜ•йҮҸеӯҗеҠӣеӯҰеҹәзЎҖзҗҶи®әж—¶жӣҙжңүеҠӣгҖӮ
3. жһ„е»әйҖ»иҫ‘пјҡдҪҝз”Ё Hadamard (H) й—ЁеңЁз¬¬дёҖдёӘжҜ”зү№дёҠеҲӣе»ә $|0\rangle$ е’Ң $|1\rangle$ зҡ„еҸ еҠ гҖӮдҪҝз”ЁдёҖзі»еҲ— CNOT (CX) й—Ёе°Ҷиҝҷз§ҚеҸ еҠ жҖҒвҖңжү©еұ•вҖқеҲ°е…¶д»–жҜ”зү№дёҠпјҢд»ҺиҖҢдә§з”ҹзә зј гҖӮ

```python
import numpy as np
from qiskit import QuantumCircuit

# 1. еҲқе§ӢеҢ– 3 дёӘйҮҸеӯҗдҪҚе’Ң 3 дёӘз»Ҹе…ёдҪҚ
n_qubits = 3
qc = QuantumCircuit(n_qubits, n_qubits)

# 2. ж–ҪеҠ й—Ёж“ҚдҪң
qc.h(0)          # еңЁз¬¬ 0 дёӘйҮҸеӯҗдҪҚдёҠж–ҪеҠ  H й—Ё
qc.cx(0, 1)      # д»Ҙ 0 дёәжҺ§еҲ¶пјҢ1 дёәзӣ®ж Үж–ҪеҠ  CNOT
qc.cx(1, 2)      # д»Ҙ 1 дёәжҺ§еҲ¶пјҢ2 дёәзӣ®ж Үж–ҪеҠ  CNOT

# 3. ж·»еҠ жөӢйҮҸж“ҚдҪң
qc.measure(range(n_qubits), range(n_qubits))

# 4. з»ҳеҲ¶з”өи·Ҝеӣҫ
print("GHZ з”өи·Ҝжһ„е»әе®ҢжҲҗпјҡ")
qc.draw()
```

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. йҖүжӢ©жЁЎжӢҹеҷЁеҗҺз«Ҝ
simulator = AerSimulator()

# 2. зј–иҜ‘е№¶иҝҗиЎҢз”өи·Ҝ (йҮҮж · 1024 ж¬Ў)
job = simulator.run(qc, shots=4096)
result = job.result()

# 3. иҺ·еҸ–жөӢйҮҸз»ҹи®Ўз»“жһң
counts = result.get_counts(qc)
f"жөӢйҮҸз»ҹи®Ўз»“жһң: {counts}"
```
