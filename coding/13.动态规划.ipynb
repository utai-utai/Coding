# 💡 专题知识点概览

::: {.callout-note}
### 核心定义

动态规划 (DP) 是一种通过把原问题分解为相对简单的子问题，从而求解复杂问题的方法。它的核心价值在于解决具有**重叠子问题**和**最优子结构**特性的问题。通过使用数组或哈希表（即 `dp` 数组）记录已解决的子问题答案（即“状态”），避免像纯暴力递归那样的重复计算。其时间复杂度通常为状态总数乘以计算每个状态所需的时间，例如一维的 $O(N)$ 或二维的 $O(N \cdot M)$。
:::

::: {.callout-tip}
### 核心要素与解题五步曲
* **状态定义**：明确 `dp` 数组（或变量）及其下标的含义。这是最难也是最关键的一步，要求什么往往就定义什么（如 `dp[i]` 表示到达第 i 阶楼梯的方法数）。
* **状态转移方程**：当前状态是怎么由前面的状态推导出来的？（如 `dp[i] = dp[i-1] + dp[i-2]`）。
* **初始化**：确定基础状态的 base case 是什么，防止越界（如 `dp[0]` 和 `dp[1]` 的值）。
* **遍历顺序**：根据状态转移方程的方向，确定是从前向后、从后向前，还是先遍历物品再遍历背包等。
* **举例推导**：手写模拟前几个状态，用于验证逻辑和排查代码错误。
:::
