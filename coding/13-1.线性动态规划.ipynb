## 📝 题目 509：斐波那契数

::: {.callout-caution}
### 📖 题目描述
**斐波那契数**（通常用 `F(n)` 表示）形成的序列称为 **斐波那契数列** 。该数列由 0 和 1 开始，后面的每一项数字都是前面两项数字的和。也就是：

* `F(0) = 0`，`F(1) = 1`
* `F(n) = F(n - 1) + F(n - 2)`，其中 n > 1

给定 `n` ，请计算 `F(n)` 。

**输入输出示例**：

* **示例 1**：
    * **输入**：`n = 2`
    * **输出**：`1`
    * **解释**：`F(2) = F(1) + F(0) = 1 + 0 = 1`

* **示例 2**：
    * **输入**：`n = 3`
    * **输出**：`2`
    * **解释**：`F(3) = F(2) + F(1) = 1 + 1 = 2`

* **示例 3**：
    * **输入**：`n = 4`
    * **输出**：`3`
    * **解释**：`F(4) = F(3) + F(2) = 2 + 1 = 3`
:::

```python
class Solution509:
    def fib(self, n: int) -> int:
        if n == 0:
            return 0
        a, b = 0, 1
        for _ in range(n-1):
            a, b = b, a + b
        return b
```

```python
#| code-fold: true
def test_509(func):
    test_cases = [
        {"desc": "边界值 n=0", "n": 0, "expected": 0},
        {"desc": "边界值 n=1", "n": 1, "expected": 1},
        {"desc": "基础值 n=2", "n": 2, "expected": 1},
        {"desc": "基础值 n=3", "n": 3, "expected": 2},
        {"desc": "基础值 n=4", "n": 4, "expected": 3},
        {"desc": "常规值 n=5", "n": 5, "expected": 5},
        {"desc": "常规值 n=6", "n": 6, "expected": 8},
        {"desc": "常规值 n=10", "n": 10, "expected": 55},
        {"desc": "较大值 n=20", "n": 20, "expected": 6765},
        {"desc": "较大值 n=30", "n": 30, "expected": 832040}
    ]

    passed = 0
    print(f"{'ID':<4} | {'结果':<6} | {'预期':<8} | {'实际':<8} | {'描述'}")
    print("-" * 60)
    for i, tc in enumerate(test_cases):
        actual = func(tc["n"])
        is_pass = actual == tc["expected"]
        status = "✅ PASS" if is_pass else "❌ FAIL"
        if is_pass:
            passed += 1
        print(f"{i+1:<4} | {status:<6} | {tc['expected']:<8} | {actual:<8} | {tc['desc']}")
    print("-" * 60)
    print(f"测试总结: 通过 {passed}/{len(test_cases)}")

# 运行测试
test_509(Solution509().fib)
```

::: {.callout-important}
### 💡 思路讲解

这道题虽然可以直接用递归来写，但递归会产生大量的重复计算。动态规划的核心思想就是**“记住已经求过的值”**。我们完全可以套用**动规五步曲**来解决：



1. **确定 dp 数组以及下标的含义**：
   - 我们定义一个一维数组 `dp`。
   - `dp[i]` 的定义为：第 `i` 个的斐波那契数值是 `dp[i]`。

2. **确定递推公式（状态转移方程）**：
   - 题目已经直接给出了规律：状态 `i` 是由前两个状态推导而来的。
   - 状态转移方程为：`dp[i] = dp[i-1] + dp[i-2]`。

3. **dp 数组如何初始化**：
   - 递推需要基础，题目明确指出 `F(0) = 0`，`F(1) = 1`。
   - 所以初始化为：`dp[0] = 0`，`dp[1] = 1`。

4. **确定遍历顺序**：
   - 从递归公式 `dp[i] = dp[i-1] + dp[i-2]` 可以看出，`dp[i]` 是依赖 `dp[i-1]` 和 `dp[i-2]` 的。
   - 因此遍历的顺序一定是从前到后遍历（即从 `i = 2` 开始递增到 `n`）。

5. **空间优化（进阶技巧）**：
   - 注意到 `dp[i]` 只依赖它前面的**两个**数字。我们其实不需要维护一个长度为 n 的数组。
   - 只需要用两个变量（例如 `prev1` 和 `prev2`）来滚动记录前两个状态，就能极大地节省内存空间。
:::

::: {.callout-note}
### 💡 复杂度分析
* **时间复杂度**：$O(N)$。我们需要从 2 遍历到 n，总共进行 n-1 次加法运算。
* **空间复杂度**：
  * 如果使用 dp 数组：$O(N)$，需要创建一个长度为 n+1 的数组。
  * 如果使用滚动变量优化：$O(1)$，只需要常数级别的额外空间。
:::

## 📝 题目 70：爬楼梯

::: {.callout-caution}
### 📖 题目描述
假设你正在爬楼梯。需要 `n` 阶你才能到达楼顶。

每次你可以爬 1 或 2 个台阶。你有多少种不同的方法可以爬到楼顶呢？

**输入输出示例**：

* **示例 1**：
    * **输入**：`n = 2`
    * **输出**：`2`
    * **解释**：有两种方法可以爬到楼顶。
      1. 1 阶 + 1 阶
      2. 2 阶

* **示例 2**：
    * **输入**：`n = 3`
    * **输出**：`3`
    * **解释**：有三种方法可以爬到楼顶。
      1. 1 阶 + 1 阶 + 1 阶
      2. 1 阶 + 2 阶
      3. 2 阶 + 1 阶
:::

```python
class Solution70:
    def climbStairs(self, n: int) -> int:
        a, b = 0, 1
        for _ in range(n):
            a, b = b, a + b
        return b
```

```python
#| code-fold: true
def test_70(func):
    test_cases = [
        {"desc": "边界值 n=1", "n": 1, "expected": 1},
        {"desc": "基础值 n=2", "n": 2, "expected": 2},
        {"desc": "基础值 n=3", "n": 3, "expected": 3},
        {"desc": "基础值 n=4", "n": 4, "expected": 5},
        {"desc": "常规值 n=5", "n": 5, "expected": 8},
        {"desc": "常规值 n=6", "n": 6, "expected": 13},
        {"desc": "常规值 n=10", "n": 10, "expected": 89},
        {"desc": "较大值 n=20", "n": 20, "expected": 10946},
        {"desc": "较大值 n=30", "n": 30, "expected": 1346269},
        {"desc": "极限值 n=45", "n": 45, "expected": 1836311903} # LeetCode 这题最大一般测到 45
    ]

    passed = 0
    print(f"{'ID':<4} | {'结果':<6} | {'预期':<10} | {'实际':<10} | {'描述'}")
    print("-" * 65)

    for i, tc in enumerate(test_cases):
        actual = func(tc["n"])
        is_pass = actual == tc["expected"]
        status = "✅ PASS" if is_pass else "❌ FAIL"
        if is_pass:
            passed += 1

        print(f"{i+1:<4} | {status:<6} | {tc['expected']:<10} | {actual:<10} | {tc['desc']}")

    print("-" * 65)
    print(f"测试总结: 通过 {passed}/{len(test_cases)}")

# 运行测试
test_70(Solution70().climbStairs)
```

::: {.callout-important}
### 💡 思路讲解

初看这道题可能会想穷举所有的组合，但随着台阶数增加，组合数会呈指数级爆炸。我们用“动规五步曲”来剥开它的外衣：



1. **确定 dp 数组以及下标的含义**：
   - 定义一维数组 `dp`。
   - `dp[i]` 表示：爬到第 `i` 层楼梯，一共有 `dp[i]` 种不同的方法。

2. **确定递推公式（状态转移方程）**：
   - 仔细思考：你是怎么到达第 `i` 层楼梯的？
   - 因为每次只能爬 1 阶或 2 阶，所以到达第 `i` 层只有两个来源：
     - 从第 `i-1` 层爬 1 阶上来。
     - 从第 `i-2` 层爬 2 阶上来。
   - 所以，到达第 `i` 层的方法数，等于到达第 `i-1` 层的方法数，加上到达第 `i-2` 层的方法数。
   - 状态转移方程为：`dp[i] = dp[i-1] + dp[i-2]`。

3. **dp 数组如何初始化**：
   - 第 1 层只有 1 种方法（爬 1 阶）：`dp[1] = 1`。
   - 第 2 层有 2 种方法（爬两个 1 阶，或爬 1 个 2 阶）：`dp[2] = 2`。
   - （注意：这道题一般不讨论 `n=0` 的情况，题目给定的 `n` 是正整数。但为了数组对齐，也可以认为 `dp[0] = 1`，具体看你的代码习惯）。

4. **确定遍历顺序**：
   - 根据公式，`dp[i]` 依赖前面的状态，所以依然是从前向后遍历（从 `i = 3` 开始到 `n`）。

5. **空间优化**：
   - 和 509 题一模一样！当前状态只依赖前两个状态，完全可以用两个变量来滚动记录，不需要开辟一整个数组。
:::

::: {.callout-note}
### 💡 复杂度分析
* **时间复杂度**：$O(N)$。需要计算 $n-2$ 次，即一层循环。
* **空间复杂度**：
  * 使用 dp 数组：$O(N)$。
  * 使用滚动变量优化：$O(1)$，仅需几个常数级的整型变量。
:::

## 📝 题目 198：打家劫舍

::: {.callout-caution}
### 📖 题目描述
你是一个专业的小偷，计划偷窃沿街的房屋。每间房内都藏有一定的现金，影响你偷窃的唯一制约因素就是相邻的房屋装有相互连通的防盗系统，**如果两间相邻的房屋在同一晚上被小偷闯入，系统会自动报警**。

给定一个代表每个房屋存放金额的非负整数数组 `nums`，计算你 **不触动警报装置的情况下** ，一夜之内能够偷窃到的最高金额。

**输入输出示例**：

* **示例 1**：
    * **输入**：`nums = [1,2,3,1]`
    * **输出**：`4`
    * **解释**：偷窃 1 号房屋 (金额 = 1) ，然后偷窃 3 号房屋 (金额 = 3)。偷窃到的最高金额 = 1 + 3 = 4 。

* **示例 2**：
    * **输入**：`nums = [2,7,9,3,1]`
    * **输出**：`12`
    * **解释**：偷窃 1 号房屋 (金额 = 2), 偷窃 3 号房屋 (金额 = 9)，接着偷窃 5 号房屋 (金额 = 1)。偷窃到的最高金额 = 2 + 9 + 1 = 12 。
:::

```python
class Solution198:
    def rob(self, nums: List[int]) -> int:
        prev1, prev2 = 0, 0  # 利用一个虚拟的“第 -1 间房”
        for num in nums:
            prev1, prev2 = max(prev1, prev2 + num), prev1
        return prev1
```

```python
#| code-fold: true
def test_198(func):
    test_cases = [
        {"desc": "空数组", "nums": [], "expected": 0},
        {"desc": "只有一间房", "nums": [5], "expected": 5},
        {"desc": "只有两间房", "nums": [2, 3], "expected": 3},
        {"desc": "示例 1 (常规)", "nums": [1, 2, 3, 1], "expected": 4},
        {"desc": "示例 2 (常规)", "nums": [2, 7, 9, 3, 1], "expected": 12},
        {"desc": "全相同金额", "nums": [2, 2, 2, 2, 2], "expected": 6},
        {"desc": "递增金额", "nums": [1, 2, 3, 4, 5], "expected": 9},
        {"desc": "交替大金额", "nums": [10, 1, 1, 10], "expected": 20},
        {"desc": "需跳过两间房", "nums": [2, 1, 1, 2], "expected": 4} # 偷头和尾
    ]

    passed = 0
    print(f"{'ID':<4} | {'结果':<6} | {'预期':<4} | {'实际':<4} | {'描述'}")
    print("-" * 65)

    for i, tc in enumerate(test_cases):
        actual = func(list(tc["nums"]))
        is_pass = actual == tc["expected"]
        status = "✅ PASS" if is_pass else "❌ FAIL"
        if is_pass:
            passed += 1
        print(f"{i+1:<4} | {status:<6} | {tc['expected']:<4} | {actual:<4} | {tc['desc']}")
    print("-" * 65)
    print(f"测试总结: 通过 {passed}/{len(test_cases)}")

# 运行测试
test_198(Solution198().rob)
```

::: {.callout-important}
### 💡 思路讲解

这道题是动态规划中“选与不选”模型的绝对经典。相比于爬楼梯，这里多了一个约束条件（不能偷相邻的），我们需要把这个约束转化为状态转移。

1. **确定 dp 数组以及下标的含义**：
   - 定义一维数组 `dp`。
   - `dp[i]` 表示：考虑下标为 `0` 到 `i` 的房屋，最多可以偷窃到的最高金额为 `dp[i]`。

2. **确定递推公式（状态转移方程）**：
   - 走到第 `i` 个房间时，你有两个决策：**偷** 或者 **不偷**。
     - **如果决策是“不偷”第 `i` 间房**：那么你的最高金额就停留在前 `i-1` 间房的最大收益，即 `dp[i] = dp[i-1]`。
     - **如果决策是“偷”第 `i` 间房**：因为不能连偷两间，所以你绝对不能偷第 `i-1` 间房。那么你当前能拿到的钱，必定是前 `i-2` 间房的最大收益，加上当前房间的钱，即 `dp[i] = dp[i-2] + nums[i]`。
   - 为了拿到最高金额，我们取这两种选择的最大值：
   - **`dp[i] = max(dp[i-1], dp[i-2] + nums[i])`**。

3. **dp 数组如何初始化**：
   - `dp[0]`：只有一间房，当然是直接偷它，`dp[0] = nums[0]`。
   - `dp[1]`：有两间房，为了收益最大且不报警，只能偷其中金额较大的那间，`dp[1] = max(nums[0], nums[1])`。

4. **确定遍历顺序**：
   - `dp[i]` 依赖于 `dp[i-1]` 和 `dp[i-2]`，所以必须是从前向后遍历（从 `i = 2` 到 `len(nums) - 1`）。

5. **空间优化**：
   - 仔细看状态转移方程！是不是和爬楼梯非常像？当前状态依然只依赖前**两**个状态。
   - 我们同样可以用两个变量（比如 `prev1` 代表 `dp[i-1]`，`prev2` 代表 `dp[i-2]`）来滚动推进，将空间复杂度降到 $O(1)$！
:::

::: {.callout-note}
### 💡 复杂度分析
* **时间复杂度**：$O(N)$。其中 $N$ 是数组 `nums` 的长度。只需要遍历一遍数组即可。
* **空间复杂度**：
  * 使用 dp 数组：$O(N)$，需要开辟长度为 $N$ 的数组。
  * 使用滚动变量优化：$O(1)$，仅需常数级别的额外空间。
:::
