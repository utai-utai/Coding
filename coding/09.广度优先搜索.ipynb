# 💡 专题知识点概览

::: {.callout-note}
### 核心定义
广度优先搜索 (BFS) 是一种利用队列 (Queue) 实现的搜索算法。它从起点开始，像水波纹一样层层向外扩散。其核心价值在于：在边权为 1 的图中，第一次到达目标节点时，所经过的路径长度一定是最短的。其时间复杂度通常为 $O(V + E)$，其中 $V$ 是节点数，$E$ 是边数。
:::

::: {.callout-tip}
### 典型应用场景
* 无权图最短路径：题目问“最少步数”、“最短距离”。由于 BFS 是按层扩散，第一次到达目标节点时，当前的层数 step 就是全局最优解。必须使用 visited 集合防止走回头路（死循环）。
* 层级/拓扑性质遍历：题目要求按层输出、求树的宽度、或处理有先后依赖的任务。在 while 内部使用 for i in range(queue.size()) 锁定当前层节点，确保每一层逻辑独立。
* 多源扩散：有多个起点（如：多个火源同时蔓延、多个感染源）。初始化时将所有起点一次性全部放入队列，视为“第 0 层”，后续逻辑与普通 BFS 一致。
* 状态空间搜索：将题目给出的非图形状态（如：一个单词、一个拼图板）抽象成“点”，每一次合法的变换抽象成“边”。需要对复杂状态进行序列化（如将矩阵转为字符串）以便存储到 visited 中。
* 拓扑排序 (Kahn's Algorithm)：解决任务依赖关系（如：选修课程 A 必须先学 B）。维护一个 入度数组 (In-degree)，每次将入度为 0 的节点入队。
:::
