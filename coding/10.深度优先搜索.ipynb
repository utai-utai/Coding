# 💡 DFS 专题知识点概览

::: {.callout-note}
### 核心定义
深度优先搜索 (DFS) 是一种利用 **栈 (Stack)** 或 **递归 (Recursion)** 实现的搜索算法。它从起点开始，遵循“不撞南墙不回头”的原则，尽可能深地搜索每一条分支。当达到目标或死胡同后，通过 **回溯 (Backtrack)** 返回上一个分叉口尝试其他路径。其核心价值在于遍历所有可能路径、处理嵌套结构以及解决复杂的决策空间问题。其时间复杂度通常为 $O(V + E)$。
:::



::: {.callout-tip}
### 典型应用场景

#### 1. 树型 DFS (结构遍历)
* **核心逻辑**：利用二叉树天然的递归定义（左/右子树也是二叉树）。
* **典型应用**：最大/最小深度计算、路径总和、平衡性判断、翻转二叉树。这是修炼递归“内功”和理解递归出口（Base Case）的第一站。

#### 2. 网格 DFS / Flood Fill (区域标记)
* **核心逻辑**：在二维矩阵中进行上下左右的扩散搜索。
* **典型应用**：岛屿数量、岛屿最大面积、图像渲染。核心是学会使用“方向数组”和“原地标记/Visited 集合”来防止陷入死循环。

#### 3. 经典回溯 / Backtracking (决策空间)
* **核心逻辑**：在“可能性的海洋”中通过尝试和撤销来寻找所有解。
* **典型应用**：全排列、组合总和、子集、N 皇后。最关键的操作三部曲：**选择 → 递归 → 撤销选择（回溯）**。

#### 4. 高阶搜索 (剪枝与记忆化)
* **核心逻辑**：通过逻辑手段让原本是指数级的搜索变得高效。
* **典型应用**：单词拆分、单词搜索、解数独。
    - **剪枝 (Pruning)**：提前断开注定失败的分支。
    - **记忆化 (Memoization)**：利用哈希表存储已计算状态，是通往动态规划 (DP) 的桥梁。
:::
