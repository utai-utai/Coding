## 📝 题目 96：不同的二叉搜索树

::: {.callout-caution}
### 📖 题目描述
给你一个整数 `n` ，求恰由 `n` 个节点组成且节点值从 `1` 到 `n` 互不相同的 **二叉搜索树 (BST)** 有多少种？

**输入输出示例**：

* **示例 1**：
    * **输入**：`n = 3`
    * **输出**：`5`
    * **解释**：3 个节点可以组成 5 种完全不同形态的树。

* **示例 2**：
    * **输入**：`n = 1`
    * **输出**：`1`

**⚠️ 数据范围**：
`1 <= n <= 19` （数据极小，这说明答案在普通整数范围内，不需要取模！）
:::
