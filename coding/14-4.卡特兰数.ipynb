## 📝 题目 96：不同的二叉搜索树

::: {.callout-caution}
### 📖 题目描述
给你一个整数 `n` ，求恰由 `n` 个节点组成且节点值从 `1` 到 `n` 互不相同的 **二叉搜索树 (BST)** 有多少种？返回满足题意的二叉搜索树的种数。

**输入输出示例 **：

* **示例 1 (单节点)**：
    * **输入**：`n = 1`
    * **输出**：`1`
    * **解释**：只有根节点 1 这一种树。

* **示例 2 (两个节点)**：
    * **输入**：`n = 2`
    * **输出**：`2`
    * **解释**：
      1. 1 是根，2 是右子节点。
      2. 2 是根，1 是左子节点。

* **示例 3 (三个节点 - 经典面试盘)**：
    * **输入**：`n = 3`
    * **输出**：`5`

* **示例 4 (四个节点)**：
    * **输入**：`n = 4`
    * **输出**：`14`

* **示例 5 (五个节点)**：
    * **输入**：`n = 5`
    * **输出**：`42`

**⚠️ 数据范围限制**：
`1 <= n <= 19`
*(提示：因为卡特兰数增长极其恐怖，$n=19$ 时结果已经高达 17 亿多，所以 LeetCode 把数据范围卡在了 19，保证可以用普通的 32/64 位整数存下，完全不需要取模！)*
:::

```python
class Solution96:
    def numTrees(self, n: int) -> int:
        return comb(2 * n, n) - comb(2 * n, n-1)
```

```python
#| code-fold: true
def test_96(func):
    test_cases = [
        {"desc": "单节点", "n": 1, "expected": 1},
        {"desc": "两个节点", "n": 2, "expected": 2},
        {"desc": "经典面试盘: 3个节点", "n": 3, "expected": 5},
        {"desc": "4个节点", "n": 4, "expected": 14},
        {"desc": "5个节点", "n": 5, "expected": 42},
        {"desc": "稍微变大: 10", "n": 10, "expected": 16796},
        {"desc": "极限 Boss: 19", "n": 19, "expected": 1767263190} # 17 亿！不用数学公式极容易超时
    ]

    passed = 0
    print(f"{'ID':<3} | {'结果':<6} | {'预期':<12} | {'实际':<12} | {'描述'}")
    print("-" * 70)

    for i, tc in enumerate(test_cases):
        actual = func(tc["n"])
        is_pass = actual == tc["expected"]
        status = "✅ PASS" if is_pass else "❌ FAIL"
        if is_pass:
            passed += 1

        print(f"{i+1:<3} | {status:<6} | {tc['expected']:<12} | {actual:<12} | {tc['desc']}")

    print("-" * 70)
    print(f"测试总结: 通过 {passed}/{len(test_cases)}")

# 运行测试
test_96(Solution96().numTrees)
```

::: {.callout-important}
### 💡 思路讲解

在传统的动态规划解法中，这道题的思路是：遍历 $1$ 到 $n$ 中的每一个数 $i$ 作为树的根节点。那么左子树由 $i-1$ 个节点组成，右子树由 $n-i$ 个节点组成。
总方案数 $G(n) = \sum_{i=1}^{n} G(i-1) \times G(n-i)$。

懂行的数学家一眼就能看出来，这个递推公式**正是卡特兰数 (Catalan Number) 的标准定义式**！
既然它本质上就是求第 $n$ 个卡特兰数，我们完全可以抛弃时间复杂度为 $O(N^2)$ 的动态规划，直接祭出卡特兰数的通项公式：
$$C_n = \binom{2n}{n} - \binom{2n}{n-1}$$

结合 Python 标准库中极其高效的 `math.comb`，这道题瞬间从“代码实现题”变成了“数学常识题”。
:::

::: {.callout-note}
### 💡 复杂度分析
* **时间复杂度**：$O(N)$。`math.comb(2n, n)` 在底层的 C 语言实现中，计算组合数只需要执行大约 $N$ 次乘除法操作。比起 DP 的 $O(N^2)$ 属于降维打击。
* **空间复杂度**：$O(1)$。没有任何多余的数组开销，只需要几个常数级别的变量存放计算结果，远胜于 DP 需要开辟长度为 $N$ 的数组。
:::

## 📝 题目 1259：不相交的握手

::: {.callout-caution}
### 📖 题目描述
偶数个人站成一个圆，总人数为 `num_people`。每个人与除自己外的一个人握手，所以总共会有 `num_people / 2` 次握手。
将这些人从 `0` 到 `num_people - 1` 按顺时针编号。

请你返回所有握手方案的数量，要求：**没有任何两人的握手线索在圆内交叉**。
由于结果可能很大，请你返回答案对 $10^9 + 7$ 取余后的结果。

**输入输出示例**：

* **示例 1**：
    * **输入**：`num_people = 2`
    * **输出**：`1`
    * **解释**：只有两个人，只能互相握手，1 种方案。

* **示例 2**：
    * **输入**：`num_people = 4`
    * **输出**：`2`
    * **解释**：
      - 方案一：0和1握手，2和3握手。
      - 方案二：0和3握手，1和2握手。
      *(注意：如果是0和2，1和3握手，两条线就会在圆心交叉，这是非法的！)*

* **示例 3**：
    * **输入**：`num_people = 6`
    * **输出**：`5`
:::

```python
class Solution1259:
    def numberOfWays(self, numPeople: int) -> int:
        return (comb(numPeople, numPeople // 2) - comb(numPeople, numPeople // 2 - 1)) % (10 ** 9 + 7)
```

```python
#| code-fold: true
def test_1259(func):
    test_cases = [
        {"desc": "示例 1: 2人 (1对)", "numPeople": 2, "expected": 1},
        {"desc": "示例 2: 4人 (2对)", "numPeople": 4, "expected": 2},
        {"desc": "示例 3: 6人 (3对)", "numPeople": 6, "expected": 5},
        {"desc": "8人 (4对)", "numPeople": 8, "expected": 14},
        {"desc": "10人 (5对)", "numPeople": 10, "expected": 42},
        {"desc": "20人 (10对)", "numPeople": 20, "expected": 16796},
        {"desc": "50人 (25对)", "numPeople": 50, "expected": 946367425},
        {"desc": "中等规模: 100人", "numPeople": 100, "expected": 265470434}, # 已经需要取模了
        {"desc": "大规模：500人", "numPeople": 500, "expected": 217193473},
        {"desc": "大厂极限压测: 1000人 (Hard 难度满级)", "numPeople": 1000, "expected": 591137401}
    ]

    passed = 0
    print(f"{'ID':<3} | {'结果':<6} | {'预期':<10} | {'实际':<10} | {'描述'}")
    print("-" * 65)

    for i, tc in enumerate(test_cases):
        actual = func(tc["numPeople"])
        is_pass = actual == tc["expected"]
        status = "✅ PASS" if is_pass else "❌ FAIL"
        if is_pass:
            passed += 1

        print(f"{i+1:<3} | {status:<6} | {tc['expected']:<10} | {actual:<10} | {tc['desc']}")

    print("-" * 65)
    print(f"测试总结: 通过 {passed}/{len(test_cases)}")

# 运行测试
test_1259(Solution1259().numberOfWays)
```

::: {.callout-important}
### 💡 思路讲解

这道题表面上是一道“圆桌连线”的几何题，但只要你试着去拆解它，就会发现它处处都在暗示卡特兰数。

**核心推导过程：**

1. **确立基准与分割**：
   总共有 `2n` 个人（即 $n$ 对）。我们随便挑一个人作为基准，比如 `0` 号。
   `0` 号必须找一个人握手。假设他找了 `k` 号握手。
2. **不交叉的铁律 (奇偶性拦截)**：
   一旦 `0` 和 `k` 连线，整个圆桌就被这条线劈成了两半：左半边和右半边。
   为了保证线绝对不交叉，左半边的人只能和左半边的人握手，右半边同理！
   这意味着：**左半边和右半边的人数，都必须是偶数！**（否则肯定有个单身狗要跨过中心线去握手，这就交叉了）。
3. **推导递推公式**：
   假设左半边有 `2i` 个人（即 $i$ 对），那么右半边自然有 `2(n - 1 - i)` 个人（即 $n - 1 - i$ 对）。
   左半边的握手方案数是 $G(i)$，右半边的方案数是 $G(n - 1 - i)$。
   总方案数就是把所有的 $i$ 遍历一遍加起来：
   $$G(n) = \sum_{i=0}^{n-1} G(i) \times G(n-1-i)$$
4. **终极降维**：
   懂行的人看到这个递推公式，DNA 直接动了——**这百分之百就是卡特兰数的标准定义式！** 就像上一题“二叉搜索树的左右子树”一样！
   既然本质是求第 $n$ 个卡特兰数，我们直接抛弃 $O(N^2)$ 的动态规划，祭出终极减法公式：
   $$C_n = \binom{2n}{n} - \binom{2n}{n-1}$$
   最后将结果对 $10^9 + 7$ 取模即可。
:::

::: {.callout-note}
### 💡 复杂度分析
* **时间复杂度**：$O(N)$。在 Python 中，`math.comb` 底层是用 C 语言优化过的循环，计算 $\binom{2N}{N}$ 大约需要 $O(N)$ 次乘除法。比起 DP 解法的 $O(N^2)$，这是绝对的降维打击。
* **空间复杂度**：$O(1)$。我们没有开辟任何 dp 数组，只用了常数级别的变量来存放最后计算出的大整数。
:::
