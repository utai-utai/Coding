```python
# dijstra
# 743
import heapq


def networkDelayTime(times: list[list[int]], n: int, k: int) -> int:
    graph = [[] for _ in range(n + 1)]
    for u, v, w in times:
        graph[u].append((v, w))
    print(graph)

    # Step 2: 初始化
    # 记录从 k 到任意点 i 的最短时间
    dist = {i: float('inf') for i in range(1, n + 1)}
    dist[k] = 0

    # 最小堆：(当前耗时, 节点ID)
    pq = [(0, k)]
    while pq:
        t, u = heapq.heappop(pq)
        if dist[u] < t:
            continue
        for v, w in graph[u]:
            if dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))
    max_wait = max(dist.values())
    return max_wait if max_wait != float('inf') else -1

times = [[2,1,1],[2,3,1],[3,4,1]]
n = 4
k = 2
networkDelayTime(times, n, k)
```

```python
#787
def findCheapestPrice(n: int, flights: list[list[int]], src: int, dst: int, k: int) -> int:
    prices = [float('inf')] * n
    prices[src] = 0

    for _ in range(k + 1):
        temp_prices = prices.copy()
        for u, v, cost in flights:
            if prices[u] == float('inf'):
                continue
            if prices[u] + cost < temp_prices[v]:
                temp_prices[v] = prices[u] + cost
        prices = temp_prices

    return prices[dst] if prices[dst] != float('inf') else -1
```

```python
#1334
def findTheCity(n: int, edges: list[list[int]], distanceThreshold: int) -> int:
    dist = [[float('inf')] * n for _ in range(n)]
    for i in range(n):
        dist[i][i] = 0
    for u, v, w in edges:
        dist[u][v] = w
        dist[v][u] = w
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
    min_count = float('inf')
    result_city = -1
    for i in range(n):
        count = 0
        for j in range(n):
            if i == j:
                continue
            if dist[i][j] <= distanceThreshold:
                count += 1
        if count <= min_count:
            min_count = count
            result_city = i
    return result_city

n = 4
edges = [[0,1,3],[1,2,1],[1,3,4],[2,3,1]]
threshold = 4
findTheCity(n, edges, threshold)
```

```python
#1334(dijkstra)
def findTheCity_dijkstra(n: int, edges: list[list[int]], distanceThreshold: int) -> int:
    # graph[u] = [(v, weight), ...]
    graph = [[] for _ in range(n)]
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))

    def dijkstra(start_node):
        pq = [(0, start_node)]
        dist = {i: float('inf') for i in range(n)}
        dist[start_node] = 0

        while pq:
            d, u = heapq.heappop(pq)
            if d > dist[u]:
                continue
            for v, d in graph[u]:
                if dist[v] > dist[u] + d:
                    dist[v] = dist[u] + d
                    heapq.heappush(pq, (dist[v], v))

        count = 0
        for i in range(n):
            if i == start_node:
                continue
            if dist[i] <= distanceThreshold:
                count += 1
        return count

    min_reachable = float('inf')
    res_city = -1
    for i in range(n):
        reachable = dijkstra(i)
        if reachable <= min_reachable:
            min_reachable = reachable
            res_city = i
    return res_city

n = 4
edges = [[0,1,3],[1,2,1],[1,3,4],[2,3,1]]
threshold = 4
findTheCity_dijkstra(n, edges, threshold)
```

```python
import random, time

def generate_graph(n, density):
    """
    生成随机无向图
    n: 节点数
    density: 密度 (0.0 - 1.0)，1.0 表示全连接图
    """
    edges = []
    # 为了保证图尽量连通，先生成一条环
    for i in range(n):
        u, v = i, (i + 1) % n
        w = random.randint(1, 10)
        edges.append([u, v, w])

    # 随机添加边
    max_edges = n * (n - 1) // 2
    target_edges = int(max_edges * density)

    current_edges = len(edges)
    while current_edges < target_edges:
        u = random.randint(0, n - 1)
        v = random.randint(0, n - 1)
        if u != v:
            w = random.randint(1, 10)
            edges.append([u, v, w])
            current_edges += 1
    return edges

def run_benchmark(name, n, density):
    print(f"\n--- 测试场景: N={n}, 密度={density*100}% ---")
    edges = generate_graph(n, density)
    threshold = 10

    # 跑 Floyd
    start = time.time()
    res_f = findTheCity(n, edges, threshold)
    time_f = time.time() - start
    print(f"Floyd    耗时: {time_f:.4f} 秒 (结果: {res_f})")

    # 跑 Dijkstra
    start = time.time()
    res_d = findTheCity_dijkstra(n, edges, threshold)
    time_d = time.time() - start
    print(f"Dijkstra 耗时: {time_d:.4f} 秒 (结果: {res_d})")

    if res_f == res_d:
        print("✅ 结果一致")
    else:
        print("❌ 结果不一致！")

# --- 4. 开始运行 ---
if __name__ == "__main__":
    # Case A: 较小规模，极度稠密 (Floyd 的舒适区)
    # N=200，意味着 Floyd 要跑 8,000,000 次计算
    run_benchmark("稠密图", n=200, density=0.8)

    # Case B: 较大规模，极度稀疏 (Dijkstra 的绝对统治区)
    # N=500，Floyd 要跑 1.25 亿次计算 (Python 会很慢)，但 Dijkstra 只需要处理很少的边
    run_benchmark("稀疏图", n=500, density=0.02)
```
