# 💡 专题知识点概览

::: {.callout-note}
### 核心定义
**最短路径 (Shortest Path)** 旨在寻找图中两点间权值之和最小的路径，是导航、网络路由等问题的底层逻辑。**最小生成树 (Minimum Spanning Tree)** 则是在无向连通图中，找出一棵包含所有顶点且边权之和最小的树，核心在于以最低成本实现“全连接”。
:::

::: {.callout-tip}
### 典型应用场景

#### 1. Dijkstra 算法 (单源最短路 - 核心地位)
* **核心逻辑**：基于**贪心策略**，通过优先队列（Heap）每次选取距离起点最近的节点，更新其邻居的距离（松弛操作）。
* **典型应用**：
    * **基础模板**：解决非负权重的点对点最短路。
    * **限制路权**：如路径上必须满足特定的体力消耗或带宽限制。
    * **路径恢复**：通过 `parent` 数组记录前驱节点，从而完整打印出最短路径。

#### 2. Bellman-Ford & SPFA (负权与中转限制)
* **核心逻辑**：通过对所有边进行多轮松弛来逼近最短路。SPFA 是其队列优化版，效率更高。

* **典型应用**：
    * **负权图**：处理包含负数权重边的图（Dijkstra 在此失效）。
    * **步数限制**：解决“最多经过 K 站中转”的问题，利用其按“轮”更新的特性限制路径长度。

#### 3. Floyd-Warshall 算法 (多源最短路 - DP思想)
* **核心逻辑**：典型的动态规划应用。三层 `for` 循环遍历中间点、起点和终点，判断是否可以通过中间点缩短距离。

* **典型应用**：
    * **全源查找**：一次性计算出图中任意两点间的最短距离。
    * **闭包传递**：判断图中节点间的连通性。

#### 4. Kruskal & Prim (最小生成树 - 成本精算)
* **核心逻辑**：
    * **Kruskal**：按边权排序，配合**并查集**判断加边是否成环，适合稀疏图。
    * **Prim**：从点集出发，不断吸纳距离当前树最近的点，适合稠密图。
* **典型应用**：
    * **基础基建**：以最低成本连接所有城市的电网、光缆或水管铺设。
    * **双解互证**：通过同一道题用两种算法实现，深度理解“加边”与“加点”的逻辑差异。
:::
