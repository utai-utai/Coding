# 💡 专题知识点概览

::: {.callout-note}
### 核心定义
数论在算法中主要研究**整数的性质**（如质数、约数、模运算），是处理大数溢出和提高计算效率的关键。组合数学则侧重于**计数问题**（如排列、组合、路径总数），核心在于通过公式或递推逻辑，将指数级的搜索空间降低为常数级或对数级的数学计算。
:::

::: {.callout-tip}
### 典型应用场景

#### 1. 快速幂与取模运算 (大数处理的基石)
* **核心逻辑**：利用二进制拆分思想，将 $O(n)$ 的连乘优化为 $O(\log n)$。在取模世界里，通过分步取模防止数值溢出。
* **典型应用**：
    * **大功率计算**：求 $x^n \pmod m$。
    * **矩阵快速幂**：将斐波那契数列等线性递推的复杂度从 $O(n)$ 降至 $O(\log n)$。

#### 2. 质数筛法与 GCD (基础算术手术刀)
* **核心逻辑**：
    * **埃氏筛 (Sieve of Erotosthenes)**：通过标记倍数，批量剔除合数，快速获取区间质数表。
    * **辗转相除法 (GCD)**：利用 `gcd(a, b) = gcd(b, a % b)` 快速求最大公约数。
* **典型应用**：判定大批量质数、求最小公倍数 (LCM)、处理分数的约分问题。

#### 3. 乘法逆元与组合数 (取模世界的除法)
* **核心逻辑**：在模 $P$（$P$ 为质数）的环境下，除以 $a$ 等价于乘以 $a^{P-2}$（费马小定理）。
* **典型应用**：
    * **大数组合计数**：求解 $\binom{n}{k} \pmod{10^9+7}$ 时，处理分母部分的除法。
    * **概率/期望题**：面试中涉及分数结果取模的终极解决方案。

#### 4. 卡特兰数 (Catalan Number - 计数的万能模板)
* **核心逻辑**：一类特定的递归结构，公式为 $C_n = \frac{1}{n+1}\binom{2n}{n}$。
* **典型应用**：
    * **合法路径/括号**：$n$ 对括号的有效组合、出栈序列总数。
    * **树形结构计数**：$n$ 个节点的二叉搜索树种类、多边形三角划分。
:::
